```python
# ^ In countplot, just giv 1 value d other will automatically b count value. But:
# In barplot we gotta give both x and y values: 1 categ, 1 numeric
sns.barplot(x='Outlet_Type',y='Item_Outlet_Sales', data=df)
```

```python
# Boxplot: gives a five-number summary. also known as whisker plot
sns.boxplot(x='Outlet_Type',y='Item_Outlet_Sales', data=df)
```

```python
# ^changing color:
sns.boxplot(x='Outlet_Type',y='Item_Outlet_Sales', data=df, palette='rainbow')
```

```python
# ^to take all features together:
sns.boxplot(data=df, orient='v')
```

```python
# ^ categorizing data based on some other categories:
sns.boxplot(x='Item_Outlet_Sales',y='Outlet_Type', hue='Outlet_Location_Type',data=df)
```
